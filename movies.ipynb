```python
mv = mv.dropna(subset = ['title'])
mv['overview'] = mv['overview'].fillna('')
mv.iloc[0]['genres']
```

```python
import ast
mv['genres'] = mv['genres'].fillna('').apply(lambda x: " ".join(i['name'] for i in ast.literal_eval(x)) if isinstance(x, str) and x.startswith('[') else x)
```

```python
mv['tags'] = mv['overview'] + ' ' + mv['genres'] + ' ' + mv['tagline']
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # lower case
    text = str(text).lower()
    # remove punctuation
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    words = text.split()
    # remove stop words
    words = [word for word in words if word not in stop_words]
    # lemmatize
    words = [lemmatizer.lemmatize(word) for word in words]
    return " ".join(words)
```

```python
mv['tags'] = mv['tags'].apply(preprocess_text)
```

```python
mv['tags'][0]
```

```python
mv = mv.reset_index(drop=True)
```

```python
indices = pd.Series(mv.index, index=mv['title'])
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(ngram_range=(1, 2), stop_words='english', max_features=50000)
```

```python
tfidf_matrix = tfidf.fit_transform(mv['tags'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n = 10):
    if title not in indices:
        return ['Movie not found']
    idx = indices[title]
    sim_scores = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
    similar_idx = sim_scores.argsort()[::-1][1:n+1]
    return mv['title'].iloc[similar_idx]
```

```python
recommend('The Dark Knight Rises')
```

```python
import pickle
mv.to_pickle('mv.pkl')
pickle.dump(indices, open('indices.pkl', 'wb'))
pickle.dump(tfidf_matrix, open('tfidf_matrix.pkl', 'wb'))
pickle.dump(tfidf, open('tfidf.pkl', 'wb'))
```
